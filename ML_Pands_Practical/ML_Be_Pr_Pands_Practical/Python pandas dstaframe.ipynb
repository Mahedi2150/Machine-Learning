## How to creat python pandas dataframe

```python
import pandas as pd
```

```python
empt_df = pd.DataFrame()
print(empt_df) 
```

```python
lst = ['a','b','c']
df1 = pd.DataFrame(lst)
print(df1)
```

```python
df1
```

```python
ls_of_ls = [[1,2,3],[4,5,6],[7,8,9]]
ls_of_ls
```

```python
df2 = pd.DataFrame(ls_of_ls)
df2
```

```python
dict1 = {"Id":[11,22,33,44]}
dict1
```

```python
dif3 = pd.DataFrame(dict1)
dif3
```

```python
dict2 = {"Id":[11,22,33,44],"SN":[1,2,3,4]}
dict2
```

```python
dif4 = pd.DataFrame(dict2)
dif4
```

```python
ls_dict = [{'a':1,'b':2},{'a':2,'b':3}]
ls_dict
```

```python
dfs = pd.DataFrame(ls_dict)
dfs
```

```python
ls_dict = [{'a':1,'b':2,"c":3},{'b':5,'c':2,'d':3,"e":4}]
ls_dict
```

```python
dfs = pd.DataFrame(ls_dict)
dfs
```

```python
dick_sr = {'id': pd.Series([1,2,3]),'SN':pd.Series ([111,222,333])}
dick_sr
```

```python
dfs2 = pd.DataFrame(dick_sr)
dfs2
```
