```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ml_student_age = np.random.randint(18,40,(100))
py_student_age = np.random.randint(15,45,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("Ml student age Histogram")
plt.xlabel("Student age categpry")
plt.ylabel("No. Student age")
```

```python
bins=[15,20,25,30,35,40]
plt.figure(figsize= (12,6))
plt.hist([ml_student_age,py_student_age],bins, rwidth = 0.8, histtype="bar",color = ["b","m"],label=["Ml Student","PY Student"])
#plt.hist(py_student_age,bins, rwidth = 0.8, histtype="bar",color = "g",label="PY Student")
plt.title("Ml student age Histogram",fontsize = 25)
plt.xlabel("Student age categpry",fontsize = 15)
plt.ylabel("No. Student age",fontsize = 15)
plt.legend()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[15,20,25,30,35,40]
plt.figure(figsize= (12,8))
plt.hist([ml_student_age,py_student_age],bins, rwidth = 0.8, histtype="bar",color = ["b","m"],label=["Ml Student","PY Student"])
#plt.hist(py_student_age,bins, rwidth = 0.8, histtype="bar",color = "g",label="PY Student")
plt.title("Ml student age Histogram",fontsize = 25)
plt.xlabel("Student age categpry",fontsize = 15)
plt.ylabel("No. Student age",fontsize = 15)
plt.legend()
```
