## String Operation Comperation and information

```python
import numpy as np
```

```python
ch_name = "Mahedi Hasan Bijoy"
str1 = "Daffodil International University"
```

```python
np.char.add(ch_name,str1)
```

```python
np.char.lower(ch_name)
```

```python
np.char.upper(str1)
```

```python
np.char.center(ch_name,60)
```

```python
np.char.center(str1,60,fillchar="*")
```

```python
np.char.split(ch_name)
```

```python
np.char.splitlines("hello\nworld")
```

```python
str4 = "dmy"
str5 = "hm"
np.char.join(["/",":"],[str4,str5])
```

```python
a = " hello 7 hi 7 world "
```

```python
np.char.replace(a,"7","5")
```

```python
np.char.equal(str4,str5)
```
