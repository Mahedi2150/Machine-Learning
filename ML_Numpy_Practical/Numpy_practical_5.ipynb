## Python Numoy Array Slicing(:)

```python
import numpy as np
```

```python
mx = np.arange(1,101).reshape(10,10)
mx
```

```python
mx[0,0]
```

```python
mx[0,0].ndim
```

```python
mx[0]
```

```python
mx[:,0:1]
```

```python
mx[:,0:1].ndim
```

```python
mx[1:7,2:5]
```

```python
mx[::]
```

```python
mx[:,:]
```
