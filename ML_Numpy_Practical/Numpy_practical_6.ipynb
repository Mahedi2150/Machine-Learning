## Python NumPy Array Conctination and split

```python
import numpy as np
```

```python
arr1 = np.arange(1,17).reshape(4,4)
print(arr1)
```

```python
arr2 = np.arange(17,33).reshape(4,4)
print(arr2)
```

```python
np.concatenate((arr1,arr2))
```

```python
np.concatenate((arr1,arr2),axis=1)
```

```python
np.vstack((arr1,arr2))
```

```python
np.hstack((arr1,arr2))
```

## Split()

```python
np.split(arr1,2)
```
