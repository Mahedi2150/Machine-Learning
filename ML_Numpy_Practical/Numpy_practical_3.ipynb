## Numpy function

```python
import numpy as np
```

## arange()

```python
ar_1d = np.arange(1,13)
print(ar_1d)
```

```python
even_arr = np.arange(1,13,2)
print(even_arr)
```

## linspace()

```python
np.linspace(1,5,4)
```

## reshape()

```python
ar_2d= ar_1d.reshape(2,6)
print(ar_2d)
```

```python
ar_3d= ar_1d.reshape(2,3,2)
print(ar_3d)
```

```python
ar = np.arange(1,13).reshape(2,6)
print(ar)
```
