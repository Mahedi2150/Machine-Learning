```python
import numpy as np
```

```python
mx_1 = np.array([[1,1,1],[1,1,1],[1,1,1]])
```

```python
print(mx_1)
```

```python
mx_1 = np.ones((3,4))
print(mx_1)
```

```python
mx_1 = np.ones((3,4), dtype=int)
print(mx_1)
```

```python
mx_1 = np.zeros((3,4))
print(mx_1)
```

```python
mx_1 = np.zeros((3,4),dtype=bool)
print(mx_1)
```

```python
mx_1 = np.zeros((3,4),dtype=str)
print(mx_1)
```
