## Mathematic Operation Using Numpy

```python
import numpy as np
```

```python
arr_1 = np.arange(1,10).reshape(3,3)
arr_2 = np.arange(1,10).reshape(3,3)
print(arr_1)
print(arr_2)
```

```python
arr_1 + arr_2
```

```python
np.add(arr_1,arr_2)
```

```python
arr_1 - arr_2
```

```python
np.subtract(arr_1,arr_2)
```

```python
arr_1 / arr_2
```

```python
np.divide(arr_1,arr_2)
```

```python
arr_1 * arr_2
```

```python
np.multiply(arr_1,arr_2)
```

```python
arr_1 @ arr_2
```

```python
np.dot(arr_1,arr_2)
```

```python
arr_1.max()
```

```python
arr_1.argmax()
```

```python
arr_1.max(axis=0)
```

```python
arr_1.max(axis=1)
```

```python
arr_1.min()
```

```python
arr_1.argmin()
```

```python
arr_1.min(axis=0)
```

```python
arr_1.min(axis=1)
```

```python
np.sum(arr_1)
```

```python
np.sum(arr_1,axis=0)
```

```python
np.sum(arr_1,axis=1)
```

```python
np.mean(arr_1)
```

```python
np.median(arr_1)
```

```python
np.sqrt(arr_1)
```

```python
np.std(arr_1)
```

```python
np.exp(arr_1)
```

```python
np.log(arr_1)
```

```python
np.log10(arr_1)
```
