```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("drive/My Drive/Colab Notebooks/Lab test/Lab-PerformanceBank-Refund.csv")
```

```python
df
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.Refund = le.fit_transform(df.Refund)
df.Marrital_Status = le.fit_transform(df.Marrital_Status)
df.Cheat = le.fit_transform(df.Cheat)
```

```python
df
```

```python
x = df.drop("Cheat",axis=1)
```

```python
x
```

```python
y = df.Cheat
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y, test_size=.2,random_state=42)
```

```python
print("xtrain",xtrain.shape)
print("xtest",xtest.shape)
print("ytrain",ytrain.shape)
print("ytest",ytest.shape)

ytest.value_counts()
```

```python
print(xtrain)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(xtrain,ytrain)
```

```python
pred = dt.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(ytest,pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(xtrain,ytrain)
```

```python
prediction = rf.predict(xtest)
```

```python
prediction
```

```python
accuracy_score(ytest,prediction)*100
```
