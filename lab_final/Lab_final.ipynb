```python
x = df.iloc[:,:-1]
```

```python
x
```

```python
y = df.iloc[:,6]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y, test_size=.2,random_state=1)
```

```python
xtrain
```

```python
ytrain
```

```python
df.isnull().sum()
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Dclf = DecisionTreeClassifier()
```

```python
Dclf.fit(xtrain,ytrain)
```

```python
predDCLF = Dclf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(ytest,predDCLF)*100)
```

#Random Forest,

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rclf = RandomForestClassifier()
```

```python
Rclf.fit(xtrain,ytrain)
```

```python
predRCLF = Rclf.predict(xtest)
```

```python
print(accuracy_score(ytest,predRCLF)*100)
```

#Support Vector Machine

```python
from sklearn import svm
```

```python
regr = svm.SVC()
```

```python
regr.fit(xtrain,ytrain)
```

```python
predSVC = regr.predict(xtest)
```

```python
print(accuracy_score(ytest,predSVC)*100)
```
